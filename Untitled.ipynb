```python
different_gaussian_nb = list()

for i in list(range(21))[2:]:
    
    obj = GaussianNB(features=data.iloc[:,1:],labels=data.iloc[:,0],train_cv_test_split=(0.7,0.2,0.1),apply_pca=True,n_principal_components=i)
    
    different_gaussian_nb.append(obj)
```

```python
different_gaussian_nb
```

```python
for obj in different_gaussian_nb:
    
    train,cv,test = obj.data_splitting()
```

```python
for obj in different_gaussian_nb:
    
    obj.fit(train)
```

```python
for obj in different_gaussian_nb:
    
    obj.evaluate(cv)
```
